```python
import requests

url = "https://rms.efy.com.vn"

payload = {"username": "rp_test", "password": "rp_test", "rpCode": "RP_TEST"}

headers = {"Content-Type": "application/json"}

try:
    response = requests.post(
        url + "/clients/login", json=payload, headers=headers, timeout=10
    )
    response.raise_for_status()  # sẽ raise exception nếu status_code != 200-399
    data = response.json()
    print("Đăng nhập thành công!")
    token = data.get("token")
    print(token)
    print()
except requests.exceptions.RequestException as e:
    print("Lỗi khi gọi API:", e)
```
